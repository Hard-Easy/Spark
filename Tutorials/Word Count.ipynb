```
subtitle_file = "/FileStore/tables/tutorials/video_subtitle.txt"
```

```
file_data = spark.sparkContext.textFile(subtitle_file)
```

```
file_data.take(3)
```

```
file_data.count()
```

```
words = file_data.flatMap(lambda line: line.lower().split(" "))
```

```
words.take(20)
```

```
word_with_number = words.map(lambda word: (word,1))
```

```
word_with_number.take(20)
```

```
word_count = word_with_number.reduceByKey(lambda x, y : x+y)
```

```
word_count.take(20)
```

```
word_count.count()
```

```
word_count.sortBy(lambda x: x[1], ascending=False).take(20)
```
